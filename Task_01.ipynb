```python
import matplotlib.pyplot as plt

plt.figure()
plt.hist(df['2022'], bins=10) # Changed 'Value' to '2022'
plt.xlabel("Population in 2022") # Updated label
plt.ylabel("Frequency")
plt.title("Distribution of Population in 2022") # Updated title
plt.show()
```

```python
import matplotlib.pyplot as plt

top10 = df[['Country Name', '2022']].sort_values(by='2022', ascending=False).head(10)

plt.figure()
plt.bar(top10['Country Name'], top10['2022'])
plt.xlabel("Country")
plt.ylabel("Population in 2022")
plt.title("Top 10 Countries by Population (2022)")
plt.xticks(rotation=45, ha='right')
plt.show()
```
